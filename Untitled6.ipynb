```python
import pandas as pd

beauty_df = pd.read_csv("most_used_beauty_cosmetics_products_extended.csv")
```

```python
beauty_df.head()
```

```python
income_df = pd.read_csv("Data Set-3.csv")
```

```python
income_df.head()
```

```python
beauty_df["Country_of_Origin"].unique()
```

```python
income_df["LOCATION"].unique()
```

```python
country_map = {
    "United States": "USA",
    "India": "IND",
    "South Korea": "KOR",
    "Japan": "JPN",
    "China": "CHN"
}
```

```python
beauty_df["LOCATION"] = beauty_df["Country_of_Origin"].map(country_map)
```

```python
merged_df = pd.merge(beauty_df, income_df, on="LOCATION")
```

```python
merged_df.head()
```

```python
merged_df.info()
```

```python
merged_df.shape
```

```python
merged_df.drop_duplicates(inplace=True)
```

```python
merged_df.isnull().sum()
```

```python
merged_df.fillna(0, inplace=True)
```

```python
merged_df.groupby("Country_of_Origin")["Value"].mean()
```

```python
merged_df["Category"].value_counts()
```

```python
merged_df.groupby("Country_of_Origin")["Rating"].mean()
```

```python
import matplotlib.pyplot as plt
plt.scatter(merged_df["Value"], merged_df["Rating"])
plt.xlabel("Disposable Income")
plt.ylabel("Beauty Product Rating")
plt.title("Income vs Beauty Preference")
plt.show()
```

```python
beauty_df["Category"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.title("Beauty Product Category Distribution")
plt.ylabel("")
plt.show()
```

```python
beauty_df.groupby("Country_of_Origin")["Rating"].mean().plot(
    kind="bar"
)

plt.xlabel("Country")
plt.ylabel("Average Rating")
plt.title("Average Beauty Rating by Country")
plt.show()
```

```python
beauty_df["Price_USD"].plot(
    kind="hist",
    bins=20
)

plt.title("Distribution of Beauty Product Prices")
plt.xlabel("Price USD")
plt.show()
```

```python
corr = merged_df[["Price_USD","Rating","Value"]].corr()

plt.imshow(corr)
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Correlation Between Income, Price and Rating")
plt.show()
```

```python
beauty_df.groupby("Category")["Price_USD"].mean().sort_values().plot(kind="bar")

plt.xlabel("Category")
plt.ylabel("Average Price")
plt.title("Average Price by Beauty Category")
plt.show() 
```

```python
top_products = beauty_df.sort_values("Rating", ascending=False).head(10)

plt.bar(top_products["Product_Name"], top_products["Rating"])
plt.xticks(rotation=90)
plt.ylabel("Rating")
plt.title("Top 10 Highest Rated Beauty Products")
plt.show()
```

```python
plt.figure(figsize=(10,5))
beauty_df.boxplot(column="Price_USD", by="Category", rot=90)

plt.title("Price Distribution by Category")
plt.suptitle("")
plt.xlabel("Category")
plt.ylabel("Price USD")
plt.show()
```
